# 1: Implement Binary Search

```python
def binary_search(arr, target):
    left, right = 0, len(arr) - 1

    while left <= right:
        mid_value = (left + right) // 2

        if arr[mid_value] == target:
            return mid_value  
        elif arr[mid_value] < target:
            left = mid_value + 1
        else:
            right = mid_value - 1 

    return -1  

arr = list(map(int, input("Enter the sorted list of numbers separated by spaces: ").split()))

target = int(input("Enter the nuumber we want to search: "))

result = binary_search(arr, target)
if result != -1:
    print(f"Element {target} found at index {result}.")
else:
    print(f"Element {target} not found in the list.")
```

# 2 : Implement Merge Sort

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    
    right = merge_sort(arr[mid:])

    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])

    return result

try:
    str_input = input("Enter numbers separated by spaces: ")
    list_input = list(map(int, str_input.split()))
    
    sorted_list = merge_sort(list_input)
    
    print("Sorted list:", sorted_list)
except:
    print("Invalid input. Please enter numbers separated by spaces.")
```

# 3 : Implement Quick Sort

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    else:
        chosen_element = arr[0]  
        less = [x for x in arr[1:] if x <= chosen_element]
        greater = [x for x in arr[1:] if x > chosen_element]
        return quick_sort(less) + [chosen_element] + quick_sort(greater)
    
u_input = input("Enter numbers separsted by spaces:")
list_input = [int(x)for x in u_input.split()]

sorted_list = quick_sort(list_input)
print("Sorted list:",sorted_list)
```

# 4 : Implement Insertion Sort

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

u_input = input("Enter numbers separated by spaces: ")
list_input= [int(x) for x in u_input.split()]

insertion_sort(list_input)
print("Sorted List:", list_input)
```
